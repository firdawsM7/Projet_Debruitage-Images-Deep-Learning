```python
# ============================================================
# 🧠 Image Denoising v4 — MÉMOIRE OPTIMISÉE (sans changer la taille)
# ============================================================
# SOLUTION AU CRASH MÉMOIRE :
#   ❌ Avant : charger 50000 images × 64×64 en RAM = 2.45 GB → CRASH
#   ✅ Maintenant : tf.data pipeline + génération de bruit à la volée
#      → RAM utilisée : ~200 MB max (×12 moins !)
#      → Taille des images : inchangée (64×64)
#      → GPU Tesla T4 utilisé à 100%
# ============================================================

# ──────────────────────────────────────────────────────────────
# SECTION 0 — Configuration mémoire GPU (à mettre EN PREMIER)
# ──────────────────────────────────────────────────────────────
import os
os.environ["TF_FORCE_GPU_ALLOW_GROWTH"] = "true"   # GPU alloue la mémoire progressivement

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, Model
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
import gc   # garbage collector

# Vérifier que le GPU est bien détecté
gpus = tf.config.list_physical_devices("GPU")
print(f"GPU détecté : {gpus}")
print(f"TensorFlow  : {tf.__version__}")

# Limiter la croissance mémoire GPU (évite les allocations d'un bloc)
for gpu in gpus:
    tf.config.experimental.set_memory_growth(gpu, True)

print("Configuration mémoire GPU : croissance progressive activée ✅")


# ──────────────────────────────────────────────────────────────
# SECTION 1 — Charger CIFAR-10 en uint8 (pas float32 !)
# ──────────────────────────────────────────────────────────────
#
# ASTUCE CLÉ :
#   float32 = 4 bytes/pixel  →  50000×64×64×3×4 = 2.45 GB  ← CRASH
#   uint8   = 1 byte/pixel   →  50000×32×32×3×1 = 0.15 GB  ← OK
#
# On garde les images en uint8 (0-255) et on normalise DANS le pipeline

(x_train_raw, _), (x_test_raw, _) = keras.datasets.cifar10.load_data()

# Garder en uint8 pour économiser la RAM
# x_train_raw.shape = (50000, 32, 32, 3)  dtype=uint8
print(f"RAM utilisée (uint8 32×32) : "
      f"{x_train_raw.nbytes / 1e6:.1f} MB")  # ~150 MB seulement
print(f"RAM si float32 64×64       : "
      f"{50000*64*64*3*4 / 1e6:.0f} MB")     # ~2457 MB = CRASH


# ──────────────────────────────────────────────────────────────
# SECTION 2 — Pipeline tf.data (mémoire efficace)
# ──────────────────────────────────────────────────────────────
#
# tf.data = les données sont traitées par BATCH, pas toutes en RAM
#   1. Une image est lue depuis le tableau uint8
#   2. Elle est upscalée 64×64 (uniquement pour le batch courant)
#   3. Du bruit est ajouté à la volée
#   4. Le batch est envoyé au GPU
#   → Seulement 2×batch_size images en RAM à la fois !

BATCH_SIZE  = 32    # réduit par rapport à v3 (128) → moins de RAM par batch
TARGET_SIZE = 64    # taille cible sans resize au test
NOISE_MIN   = 0.02
NOISE_MAX   = 0.40
AUTOTUNE    = tf.data.AUTOTUNE

@tf.function
def preprocess_and_noise(image):
    """
    Transformations appliquées sur CHAQUE image dans le pipeline :
    1. Cast uint8 → float32 et normalisation [0,1]
    2. Upscale 32×32 → 64×64 (bilinéaire)
    3. Ajout de bruit gaussien aléatoire
    Retourne : (image_bruitée, image_propre)
    """
    # Cast et normalisation
    image = tf.cast(image, tf.float32) / 255.0

    # Upscale 32→64 (uniquement en mémoire pour le batch courant)
    image = tf.image.resize(image, [TARGET_SIZE, TARGET_SIZE], method="bilinear")

    # Augmentation : flip horizontal aléatoire (gratuit, améliore généralisation)
    image = tf.image.random_flip_left_right(image)

    # Bruit gaussien multi-niveau
    noise_std = tf.random.uniform([], NOISE_MIN, NOISE_MAX)
    noise     = tf.random.normal(tf.shape(image), stddev=noise_std)
    noisy     = tf.clip_by_value(image + noise, 0.0, 1.0)

    return noisy, image   # (input bruité, target propre)

def make_dataset(images, shuffle=True):
    """
    Crée un tf.data.Dataset optimisé mémoire.
    Le bruit est généré à la volée à chaque epoch (data augmentation gratuite !).
    """
    ds = tf.data.Dataset.from_tensor_slices(images)
    if shuffle:
        ds = ds.shuffle(buffer_size=2000, reshuffle_each_iteration=True)
    ds = (ds
          .map(preprocess_and_noise, num_parallel_calls=AUTOTUNE)
          .batch(BATCH_SIZE)
          .prefetch(AUTOTUNE))   # précharge le batch suivant pendant l'entraînement GPU
    return ds

print("Création des datasets tf.data ...")
train_ds = make_dataset(x_train_raw, shuffle=True)
test_ds  = make_dataset(x_test_raw,  shuffle=False)
print("Datasets créés ✅  (bruit généré à la volée, mémoire minimale)")


# ──────────────────────────────────────────────────────────────
# SECTION 3 — Architecture DnCNN optimisée mémoire
# ──────────────────────────────────────────────────────────────
#
# Ajustements vs v3 pour éviter l'OOM sur le GPU :
#   • filters réduit de 64 → 48  (−25% mémoire GPU)
#   • depth  réduit de 17 → 12   (−30% mémoire GPU)
#   • Résidual learning conservé (prédit le bruit, pas l'image)

def build_dncnn_lite(depth=12, filters=48, input_shape=(None, None, 3)):
    """
    DnCNN-Lite : version allégée pour tenir dans 14 GB GPU.
    Fully convolutional → fonctionne sur n'importe quelle résolution.
    """
    inputs = keras.Input(shape=input_shape)

    # Première couche (sans BatchNorm)
    x = layers.Conv2D(filters, 3, padding="same", activation="relu")(inputs)

    # Couches intermédiaires
    for _ in range(depth - 2):
        x = layers.Conv2D(filters, 3, padding="same", use_bias=False)(x)
        x = layers.BatchNormalization()(x)
        x = layers.Activation("relu")(x)

    # Prédiction du bruit résiduel
    noise_pred = layers.Conv2D(3, 3, padding="same")(x)

    # image propre = image bruitée - bruit prédit
    output = layers.Subtract()([inputs, noise_pred])
    output = layers.Lambda(lambda t: tf.clip_by_value(t, 0.0, 1.0))(output)

    return Model(inputs, output, name="DnCNN_Lite")


model = build_dncnn_lite()
model.summary()

# Estimation de la mémoire GPU utilisée par le modèle
params     = model.count_params()
model_mb   = params * 4 / 1e6
print(f"\nParamètres   : {params:,}")
print(f"Mémoire modèle (approx) : {model_mb:.1f} MB")
print(f"GPU disponible          : ~13757 MB (Tesla T4)")


# ──────────────────────────────────────────────────────────────
# SECTION 4 — Loss combinée
# ──────────────────────────────────────────────────────────────

def gradient_loss(y_true, y_pred):
    dy_t, dx_t = tf.image.image_gradients(y_true)
    dy_p, dx_p = tf.image.image_gradients(y_pred)
    return tf.reduce_mean(tf.abs(dx_t - dx_p) + tf.abs(dy_t - dy_p))

def full_loss(y_true, y_pred):
    mse  = tf.reduce_mean(tf.square(y_true - y_pred))
    ssim = 1.0 - tf.reduce_mean(tf.image.ssim(y_true, y_pred, max_val=1.0))
    grad = gradient_loss(y_true, y_pred)
    return 0.4 * mse + 0.4 * ssim + 0.2 * grad

model.compile(
    optimizer=keras.optimizers.Adam(1e-3),
    loss=full_loss,
    metrics=["mae"]
)
print("Modèle compilé ✅")


# ──────────────────────────────────────────────────────────────
# SECTION 5 — Entraînement
# ──────────────────────────────────────────────────────────────

callbacks = [
    ReduceLROnPlateau(monitor="val_loss", factor=0.5,
                      patience=4, min_lr=1e-6, verbose=1),
    EarlyStopping(monitor="val_loss", patience=10,
                  restore_best_weights=True, verbose=1),
    # Libère la mémoire GPU entre les epochs
    keras.callbacks.LambdaCallback(
        on_epoch_end=lambda epoch, logs: gc.collect()
    )
]

history = model.fit(
    train_ds,
    epochs          = 50,
    validation_data = test_ds,
    callbacks       = callbacks,
    verbose         = 1
)
print("\nEntraînement terminé ✅")
model.save("dncnn_lite.keras")


# ──────────────────────────────────────────────────────────────
# SECTION 6 — Courbes de loss
# ──────────────────────────────────────────────────────────────

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history["loss"],     label="Train", lw=2)
plt.plot(history.history["val_loss"], label="Val",   lw=2, linestyle="--")
plt.title("Loss (MSE+SSIM+Gradient)")
plt.xlabel("Epoch"); plt.legend(); plt.grid(alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(history.history["mae"],     label="Train MAE", lw=2)
plt.plot(history.history["val_mae"], label="Val MAE",   lw=2, linestyle="--")
plt.title("MAE"); plt.xlabel("Epoch"); plt.legend(); plt.grid(alpha=0.3)

plt.tight_layout()
plt.savefig("loss_v4.png", dpi=100)
plt.show()


# ──────────────────────────────────────────────────────────────
# SECTION 7 — Visualisation des résultats
# ──────────────────────────────────────────────────────────────

# Récupérer quelques exemples du dataset test pour visualisation
sample_noisy, sample_clean = next(iter(test_ds.take(1)))
sample_noisy = sample_noisy[:8].numpy()
sample_clean = sample_clean[:8].numpy()
sample_denoised = model.predict(sample_noisy, verbose=0)

fig, axes = plt.subplots(3, 8, figsize=(16, 6))
for col in range(8):
    for row, (imgs, title) in enumerate(zip(
        [sample_noisy, sample_denoised, sample_clean],
        ["Bruité", "Débruité", "Original"]
    )):
        axes[row, col].imshow(np.clip(imgs[col], 0, 1))
        axes[row, col].axis("off")
        if col == 0:
            axes[row, col].set_ylabel(title, fontsize=9,
                                       rotation=0, labelpad=65, va="center")

plt.suptitle("DnCNN-Lite v4 — Résultats (images 64×64)", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.savefig("results_v4.png", dpi=120)
plt.show()


# ──────────────────────────────────────────────────────────────
# SECTION 8 — Test image externe HAUTE RÉSOLUTION
# ──────────────────────────────────────────────────────────────
#
# DnCNN est fully convolutional → traite l'image à sa taille originale
# Aucun resize, aucune perte de détails

from google.colab import files
from PIL import Image
import io

def test_high_res(model, max_dim=768):
    """
    Débruite une image uploadée à sa RÉSOLUTION ORIGINALE.
    Gestion mémoire : limite optionnelle pour les très grandes images.
    """
    print("📂 Upload ton image bruitée ...")
    uploaded = files.upload()
    if not uploaded:
        print("⚠️  Aucun fichier uploadé.")
        return

    filename  = list(uploaded.keys())[0]
    img       = Image.open(io.BytesIO(uploaded[filename])).convert("RGB")
    orig_size = img.size
    print(f"   → Taille originale : {orig_size[0]}×{orig_size[1]} px")

    # Limite mémoire optionnelle
    if max_dim and max(orig_size) > max_dim:
        scale = max_dim / max(orig_size)
        img   = img.resize((int(orig_size[0]*scale), int(orig_size[1]*scale)), Image.LANCZOS)
        print(f"   → Redimensionnée  : {img.size[0]}×{img.size[1]} px")
    else:
        print("   → Résolution originale conservée ✅")

    # Prétraitement
    img_arr   = np.array(img, dtype="float32") / 255.0
    img_batch = np.expand_dims(img_arr, axis=0)   # (1, H, W, 3)

    # Inférence
    print("   → Inférence en cours ...")
    output    = model.predict(img_batch, verbose=0)
    denoised  = np.clip(output[0], 0.0, 1.0)

    # Métriques
    psnr_val = 10 * np.log10(1.0 / (np.mean((img_arr - denoised)**2) + 1e-10))
    noise_rm = np.mean(np.abs(img_arr - denoised))
    print(f"   → PSNR              : {psnr_val:.2f} dB")
    print(f"   → Bruit supprimé    : {noise_rm:.4f}")

    # Affichage
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))
    fig.suptitle(
        f"{filename}  |  {img_arr.shape[1]}×{img_arr.shape[0]} px  "
        f"|  PSNR: {psnr_val:.1f} dB",
        fontsize=12
    )
    axes[0].imshow(img_arr);    axes[0].set_title("Bruité (Input)");   axes[0].axis("off")
    axes[1].imshow(denoised);   axes[1].set_title("Débruité (Output)"); axes[1].axis("off")
    diff = np.clip(np.abs(img_arr - denoised) * 5, 0, 1)
    axes[2].imshow(diff, cmap="hot")
    axes[2].set_title("Bruit supprimé (×5)"); axes[2].axis("off")

    plt.tight_layout()
    plt.savefig("denoised_v4.png", dpi=150, bbox_inches="tight")
    plt.show()

    # Sauvegarder et télécharger l'image débruitée
    Image.fromarray((denoised * 255).astype(np.uint8)).save("denoised_output_v4.png")
    files.download("denoised_output_v4.png")
    print("\n✅ Fichiers sauvegardés et téléchargés !")
    print("   • denoised_v4.png         (comparaison)")
    print("   • denoised_output_v4.png  (image débruitée seule)")

    # Libérer la mémoire
    del img_batch, output
    gc.collect()


test_high_res(model, max_dim=768)
```
